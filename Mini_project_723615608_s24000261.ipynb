```python
from collections import deque  # Import deque (double-ended queue) for efficient add/remove operations

class Vehicle:
    # constructor: initializes attributes common to all vehicles
    def __init__(self, vehicle_id, vehicle_type):
        self.vehicle_id = vehicle_id      # Unique identifier for each vehicle
        self.vehicle_type = vehicle_type  # Type of vehicle
        self.is_hired = False             # Checking whether the vehicle is currently hired

    # Method to return a string version of the vehicle(for each display)
    def __str__(self):
        status = "Hired" if self.is_hired else "Available"                # Determine whether vehicle available or hired
        return f"{self.vehicle_type} - ID: {self.vehicle_id} - {status}"  # format string


class Car(Vehicle):  # Car inherits from vehicle
    def __init__(self, vehicle_id, max_passengers, ac):
        super().__init__(vehicle_id, "Car")   # Call parent class constructor with vehicle type "car"
        self.max_passengers = max_passengers  # Maximum passenger capacity
        self.ac = ac                          # Boolean indicating if the car has AC or not

    # Method to check if car matches passenger and AC requirements
    def matches(self, passengers, ac):
        return self.max_passengers == passengers and self.ac == ac  # Returns True if both match

    def __str__(self):
        ac_status = "AC" if self.ac else "Non-AC"  # Convert boolean to readable string
        status = "Hired" if self.is_hired else "Available"  # Include current hire status
        return f"Car(ID: {self.vehicle_id}, Seats: {self.max_passengers}, {ac_status}, {status})"   # Display car info


class Van(Vehicle):  # Van inherits from vehicle
    def __init__(self, vehicle_id, max_passengers, ac):
        super().__init__(vehicle_id, "Van")   # Call parent constructor
        self.max_passengers = max_passengers  # Max passenger capacity
        self.ac = ac                          # AC availability (True/False)

    # Method to check if van meets user requirements
    def matches(self, passengers, ac):
        return self.max_passengers == passengers and self.ac == ac  # Match on both attributes

    # String method for van info
    def __str__(self):
        ac_status = "AC" if self.ac else "Non-AC"   # Convert boolean to readable text
        status = "Hired" if self.is_hired else "Available"  # Include hire status
        return f"Van(ID: {self.vehicle_id}, Seats: {self.max_passengers}, {ac_status}, {status})"  # Show information


class ThreeWheeler(Vehicle):  # Inherits from vehicle
    def __init__(self, vehicle_id):
        super().__init__(vehicle_id, "Three Wheeler")  # Call base constructor
        self.max_passengers = 3  # Fixed capacity for all three wheelers

    # Method to check the passenger requirement is <=3
    def matches(self, passengers):
        return passengers <= 3

    # String representation of a three wheeler
    def __str__(self):
        status = "Hired" if self.is_hired else "Available"  # Include hire status
        return f"Three Wheeler(ID: {self.vehicle_id}, Seats: 3, {status})"  #Show information


class Truck(Vehicle):  # Truck inherits from vehicle
    def __init__(self, vehicle_id, size_ft):
        super().__init__(vehicle_id, "Truck")  # Initialize with vehicle type "Truck"
        self.size_ft = size_ft   # Truck size (in feet)

    # Check if truck matches the requested size
    def matches(self, size_ft):
        return self.size_ft == size_ft

    # String version of the Truck
    def __str__(self):
        status = "Hired" if self.is_hired else "Available"  # Include hire status
        return f"Truck(ID: {self.vehicle_id}, Size: {self.size_ft}ft, {status})"


class Lorry(Vehicle):  # Lorry inherits from vehicle
    def __init__(self, vehicle_id, max_load):
        super().__init__(vehicle_id, "Lorry")  # Initialize with vehicle type "Lorry"
        self.max_load = max_load  # Load capacity in kgs

    # Match load requirements
    def matches(self, load):
        return self.max_load == load

    # String version of a Lorry
    def __str__(self):
        status = "Hired" if self.is_hired else "Available"  # Include hire status
        return f"Lorry(ID: {self.vehicle_id}, Load: {self.max_load}kg, {status})"


class CabService:
    def __init__(self):
        # Use a dictionary with vehicle type as key, and deque (queue) as value
        self.vehicles = {
            "Car": deque(),            # Queue for cars
            "Van": deque(),            # Queue for Vans
            "Three Wheeler": deque(),  # Queue for three wheelers
            "Truck": deque(),          # Queue for Trucks
            "Lorry": deque()           # Queue for lorries
        }

    def vehicle_exists(self, vehicle_id):
        """Check if a vehicle with the same ID already exists."""
        for queue in self.vehicles.values():  # Go through each every category
            for v in queue:                   # Go through each vehicle in the category
                if v.vehicle_id == vehicle_id:  # Compare the vehicle ID
                    return True                # Return True if found
        return False    # Return False if not found

    def add_vehicle(self, vehicle):
        if self.vehicle_exists(vehicle.vehicle_id):  # Prevent duplicate IDs
            print(f"Error: Vehicle with ID '{vehicle.vehicle_id}' already exists.")
            return
        # Add the vehicle to its respective queue
        self.vehicles[vehicle.vehicle_type].append(vehicle)
        print(f"Added {vehicle}")   # Confirm addition

    def remove_vehicle(self, vehicle_id):
        # Loop through all queues
        for category, queue in self.vehicles.items():
            for v in queue:
                if v.vehicle_id == vehicle_id: # Find matching ID
                    queue.remove(v)  # Remove from queue
                    print(f"Removed {v}")  # Confirm removal
                    return
        print("Vehicle not found")  # If not found display message

    def assign_job(self, vehicle_type, **kwargs):
        queue = self.vehicles.get(vehicle_type) # Get the relevant queue
        if not queue:  # If vehicle type invalid
            print("Invalid vehicle type")
            return

        # Search for available vehicle
        for v in queue:
            if not v.is_hired:  # Only assign if vehicle is available
                # Check vehicle type-specific conditions
                if (
                    # If the customer requested a car, check passenger count and AC requirement
                    (vehicle_type == "Car" and v.matches(kwargs["passengers"], kwargs["ac"])) or
                    # If it's a van, check the passengers + Ac condition
                    (vehicle_type == "Van" and v.matches(kwargs["passengers"], kwargs["ac"])) or
                    # If it's a three Wheeler , check the requested passenger within capacity (3)
                    (vehicle_type == "Three Wheeler" and v.matches(kwargs["passengers"])) or
                    # If it's a truck, match the required size
                    (vehicle_type == "Truck" and v.matches(kwargs["size_ft"])) or
                    # If it's a lorry, match the required load capacity
                    (vehicle_type == "Lorry" and v.matches(kwargs["load"]))
                ):
                    v.is_hired = True  # Mark as hired
                    print(f"Assigned {v} for the job")  # Show assignment
                    return
        print("No available vehicle matching the requirements")  # If no match found

    def release_vehicle(self, vehicle_id):
        # Loop through all vehicles
        for queue in self.vehicles.values():
            for v in queue:
                if v.vehicle_id == vehicle_id:
                    if v.is_hired:  # Check if vehicle was hired
                        v.is_hired = False  # Release it
                        print(f"Vehicle released: {v}")
                        return
                    else:
                        print("Vehicle was not hired")  # Already free
                        return
        print("Vehicle not found")  # If not in system

    def show_available(self):
        print("\nAvailable Vehicles:")
        # Go through each vehicle category
        for category, queue in self.vehicles.items():
            # Filter out only vehicles not hired
            available = [v for v in queue if not v.is_hired]
            if available:
                print(f"{category}s:")  # Print category name
                for v in available:
                    print(" ", v)   # Print vehicle details
            else:
                print(f"{category}s: None available")  # If none available
        print()  # Add spacing line


if __name__ == "__main__":  # Run only if file executed directly
    service = CabService()   # Create cabservice object

    # Preload some vehicles into the system
    service.add_vehicle(Car("C1", 3, True))
    service.add_vehicle(Car("C2", 3, False))
    service.add_vehicle(Car("C3", 4, True))
    service.add_vehicle(Car("C4", 4, False))
    service.add_vehicle(Van("V1", 6, True))
    service.add_vehicle(Van("V2", 6, False))
    service.add_vehicle(Van("V3", 8, True))
    service.add_vehicle(Van("V4", 8, False))
    service.add_vehicle(ThreeWheeler("T1"))
    service.add_vehicle(Truck("TR1", 7))
    service.add_vehicle(Truck("TR2", 12))
    service.add_vehicle(Lorry("L1", 2500))
    service.add_vehicle(Lorry("L2", 3500))

    while True:  # Infinite loop until exit option selected
        # Display menu
        print("\nCab Service Menu")
        print("1 Add Vehicle")
        print("2 Remove Vehicle")
        print("3 Assign Job")
        print("4 Release Vehicle")
        print("5 Show Available Vehicles")
        print("6 Exit")

        choice = input("Enter your choice: ") # Get user's menu choice

        try:
            if choice == "1":
                v_type = input("Enter vehicle type (Car Van Three Wheeler Truck Lorry): ").title()  # Get type
                vid = input("Enter vehicle ID: ")  # Get ID

                # Depending on type, collect necessary data and create object
                if v_type == "Car":
                    try:
                        passengers = int(input("Enter max passengers (3 or 4): ")) # Get passenger count
                        # Validate the entered passenger number
                        if passengers not in [3, 4]:
                            print("Invalid passenger count for Car") # Inform the passenger count is invalid
                            continue  # Go back to main menue without crashing
                        ac = input("AC (y n): ").lower() == "y" # Ask if the car has air conditioning
                        service.add_vehicle(Car(vid, passengers, ac)) # Create an object with the entered details and add it to the system
                    except ValueError:
                        print("Invalid input. Please enter numeric value for passengers.") # This block catches invalid input

                elif v_type == "Van":
                    try:
                        passengers = int(input("Enter max passengers (6 or 8): "))   # Get passenger count
                        # Validate the entered passenger number
                        if passengers not in [6, 8]:
                            print("Invalid passenger count for Van")   # Inform the passenger count is invalid
                            continue   # Go back to main menue without crashing
                        ac = input("AC (y n): ").lower() == "y"   # Ask if the Van has air conditioning
                        service.add_vehicle(Van(vid, passengers, ac))  # Create an object with the entered details and add it to the system
                    except ValueError:
                        print("Invalid input. Please enter numeric value for passengers.")  # This block catches invalid input

                elif v_type == "Three Wheeler":
                    service.add_vehicle(ThreeWheeler(vid))  # Add to the system

                elif v_type == "Truck":
                    try:
                        size = int(input("Enter size (7 or 12 ft): "))  # Get the size in feet
                        # Validate the entered size
                        if size not in [7, 12]:
                            print("Invalid size for Truck")  # Inform the size is invalid
                            continue   # Go back to main menue without crashing
                        service.add_vehicle(Truck(vid, size))  # Create an object with the entered details and add it to the system
                    except ValueError:
                        print("Invalid input. Please enter numeric value for size.")   # This block catches invalid input

                elif v_type == "Lorry":
                    try:
                        load = int(input("Enter max load (2500 or 3500 kg): "))  # Get the max load in kg
                        # Validate the entered max load
                        if load not in [2500, 3500]:
                            print("Invalid load for Lorry")  # Inform the max load is invalid
                            continue  # Go back to main menue without crashing
                        service.add_vehicle(Lorry(vid, load))  # Create an object with the entered details and add it to the system
                    except ValueError:
                        print("Invalid input. Please enter numeric value for load.")   # This block catches invalid input

                else:
                    print("Invalid type")  # Invalid entry

            elif choice == "2":  # Option 2- Remove a vehicle
                vid = input("Enter vehicle ID to remove: ")
                service.remove_vehicle(vid)

            elif choice == "3":  # Option 3- Assign a job
                v_type = input("Enter vehicle type: ").title()
                # Collect parameters depending on vehicle type
                # Check if the job is for a car or a van
                if v_type in ["Car", "Van"]:
                    try:
                        passengers = int(input("Enter passenger count: "))  # Get passenger count
                        ac = input("Need AC (y n): ").lower() == "y" # Ask the user whether AC is required
                        # Assign a vehicle for the job based on the type, passenger count, Ac requirement
                        service.assign_job(v_type, passengers=passengers, ac=ac)
                    except ValueError:
                        # This block catches the invalid input
                        print("Invalid input for passenger count.")
                elif v_type == "Three Wheeler":
                    try:
                        passengers = int(input("Enter passenger count (up to 3): "))  # Get the passenger count(upto 3)
                        # Assign a vehicle for the job based on the type, passenger count
                        service.assign_job(v_type, passengers=passengers)
                    except ValueError:
                        # This block catches the invalid input
                        print("Invalid input for passenger count.")
                elif v_type == "Truck":
                    try:
                        size = int(input("Enter size (7 or 12): "))  # Get the size in feet
                        # Assign a vehicle for the job based on the type, size
                        service.assign_job(v_type, size_ft=size)
                    except ValueError:
                        # This block catches the invalid input
                        print("Invalid input for size.")
                elif v_type == "Lorry":
                    try:
                        load = int(input("Enter load (2500 or 3500): "))   # Get the max load in Kg
                        # Assign a vehicle for the job based on the type, max load
                        service.assign_job(v_type, load=load)
                    except ValueError:
                        # This block catches the invalid input
                        print("Invalid input for load.")
                else:

                    print("Invalid vehicle type")

            elif choice == "4":  # Option 4- Release Vehicle
                vid = input("Enter vehicle ID to release: ")  # Get the vehicle ID
                service.release_vehicle(vid)

            elif choice == "5":  # Option 5- Show available vehicles
                service.show_available()

            elif choice == "6":  # Option 6- Exit
                print("Exiting Cab Service System")
                break  # Exit loop

            else:
                print("Invalid choice")  # Invalid input handler

        except Exception as e:
            print(f"An unexpected error occurred: {e}")  # Handle unexpected runtime errors safely
```
